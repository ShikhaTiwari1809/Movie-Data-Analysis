## ********* This is a data collection process and we are utlizing TheMovieDb.org APIs to fetch data for 1000 most popular movie and will perform EDA on that dataset. ********* 

### 1.Importing Libraries

```python
import requests
import pandas as pd
import warnings
warnings.filterwarnings("ignore")
```

### 2. Getting data from " https://developers.themoviedb.org/3/movies/get-popular-movies "  and fetching only Movie Id and Title and saving it to a csv file.

```python
movielist =pd.DataFrame()
for pageno in range(1,500):
    url = 'https://api.themoviedb.org/3/movie/popular?api_key=0e9842506d4cc5e2395013ebac9bd772&language=en-US&page={}'.format(pageno)
    response = requests.get(url).json()
    title=[]
    id=[]
    try:
        for i in response['results']:
            title.append(i['title'])
            id.append(i['id'])
        d = {'title':title,'id':id}
    except:
        pass
    
    df = pd.DataFrame(d)
    movielist = movielist.append(df,ignore_index=True)
```

```python
movieidlist= pd.read_csv('movielist.csv')
movieidlist
```

### 3. Saving all the Movie Id into a list which will be used to fetch Movie Details data from " https://developers.themoviedb.org/3/movies/get-movie-details " 

```python
mid =movieidlist['id'].to_list()
```

### 4. Iterating over all Movie ids and getting required movie details data and saving it to csv file.

```python
moviedata = pd.DataFrame()
for id in mid[:2001]:
    url = 'https://api.themoviedb.org/3/movie/{}?api_key=0e9842506d4cc5e2395013ebac9bd772&language=en-US'.format(id)
    response = requests.get(url,headers={"User-Agent":"Mozilla/5.0"}).json()
    budget=[]
    genres=[]
    id=[]
    overview=[]
    popularity=[]
    production_companies=[]
    release_date=[]
    revenue=[]
    title=[]
    vote_average=[]
    original_language=[]
    vote_count=[]
    try:
        budget.append(response['budget'])
        genres.append(response['genres'])
        id.append(response['id'])
        overview.append(response['overview'])
        popularity.append(response['popularity'])
        production_companies.append(response['production_companies'])
        release_date.append(response['release_date'])
        revenue.append(response['revenue'])
        title.append(response['title'])
        vote_average.append(response['vote_average'])
        original_language.append(response['original_language'])
        vote_count.append(response['vote_count'])

        d = {'id':id,
            'budget':budget,
            'genres':genres,
            'overview':overview,
            'popularity':popularity,
            'production_companies':production_companies,
            'release_date':release_date,
            'revenue':revenue,
            'title':title,
            'vote_average':vote_average,
            'original_language':original_language,
            'vote_count':vote_count}

        
        
    except:
        pass
    
    df = pd.DataFrame(d)
    moviedata = moviedata.append(df,ignore_index=True)
```

```python
moviedata.to_csv('moviedata.csv')
```

```python
mid =moviedata['id'].to_list()
```

### 5. Fetching Cast and crew data for all the movie ids from " https://developers.themoviedb.org/3/movies/get-movie-credits " and saving that to csv file which will be utilized in EDA process

```python
cast_crew = pd.DataFrame()
for id in mid:
    url = 'https://api.themoviedb.org/3/movie/{}/credits?api_key=0e9842506d4cc5e2395013ebac9bd772&language=en-US'.format(id)
    response = requests.get(url,headers={"User-Agent":"Mozilla/5.0"}).json()
    id=[]
    cast=[]
    crew=[]
    try:
        id.append(response['id'])
        cast.append(response['cast'])
        crew.append(response['crew'])
        
        d={'id':id,
          'cast':cast,
          'crew':crew}
    except:
        pass
    
    df = pd.DataFrame(d)
    cast_crew=cast_crew.append(df,ignore_index=True)
```
